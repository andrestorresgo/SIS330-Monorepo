# Confusion Matrix — Tier-1 Hemo-Scout (1D-CNN · PLETH)

**The Paranoid Watchdog** — Catches 100% of cardiovascular crashes, but panics on every sensor bump.

Evaluation set: 200 crash events · 800 safe/artifact events

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

# ── Hemo-Scout confusion matrix (from cm_hemo_scout.csv) ──────────────
#                  Predicted Safe   Predicted Crash
# Actual Safe         526              274
# Actual Crash          0              200

matrix = np.array([
    [526, 274],   # Actual Safe  → TN, FP
    [  0, 200],   # Actual Crash → FN, TP
])

row_labels = ["Actual Safe", "Actual Crash"]
col_labels = ["Predicted Safe", "Predicted Crash"]

# ── Plot ──────────────────────────────────────────────────────────────
cmap = LinearSegmentedColormap.from_list("alarm", ["#fef0f0", "#e74c3c", "#922b21"])

total = matrix.sum()
annot = np.array([
    [f"{v:,}\n({v/total*100:.1f}%)" for v in row]
    for row in matrix
])

fig, ax = plt.subplots(figsize=(6.5, 5.5))

sns.heatmap(
    matrix,
    annot=annot,
    fmt="",
    cmap=cmap,
    linewidths=2.5,
    linecolor="white",
    cbar=False,
    square=True,
    xticklabels=col_labels,
    yticklabels=row_labels,
    annot_kws={"size": 18, "weight": "bold"},
    ax=ax,
)

ax.set_title("Tier-1 Hemodinamia  (1D-CNN · PLETH)\n",
             fontsize=14, fontweight="bold", pad=16)
ax.set_ylabel("Actual", fontsize=12, fontweight="semibold")
ax.set_xlabel("Predicted", fontsize=12, fontweight="semibold")
ax.tick_params(labelsize=11)

# Quadrant labels
labels = [["TN", "FP"], ["FN", "TP"]]
for i in range(2):
    for j in range(2):
        ax.text(j + 0.5, i + 0.88, labels[i][j],
                ha="center", va="center",
                fontsize=10, color="#555555", style="italic")

fig.tight_layout()
plt.show()

# ── Metrics ───────────────────────────────────────────────────────────
tp, tn, fp, fn = 200, 526, 274, 0
print(f"Accuracy:            {(tp+tn)/total*100:.2f}%")
print(f"Precision:           {tp/(tp+fp)*100:.2f}%")
print(f"Recall (Sensitivity):{tp/(tp+fn)*100:.2f}%")
print(f"F1 Score:            {2*tp/(2*tp+fp+fn)*100:.2f}%")
print(f"False Positive Rate: {fp/(fp+tn)*100:.2f}%")
```
